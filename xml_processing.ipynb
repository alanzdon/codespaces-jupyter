```python
# Make custom function for making text human- (and pipeline)-readable
def norm_and_strip(s: str) -> str:
    normalized = unicodedata.normalize('NFKC', s)
    stripped = re.sub(r'\n\s+', ' ', normalized).strip()
    return stripped
```

```python
# Start scaling project to entire corpus
import os
folder_files = os.listdir('mercure-galant/xml')
print("This Mercure Galant corpus contains {len_folder} tomes.".format(len_folder=len(folder_files)))
```

```python
# Put tomes in chronological order
folder_files.sort()

# Start by writing first tome to CSV to include headers
tome = folder_files[0]

# Read first tome XML and parse into soup object
with open('mercure-galant/xml/'+tome, 'r') as f:
    soup = bs.BeautifulSoup(f, 'xml')

bibls = [] # Bibliography list
titles = [] # Title list
articles = [] # Article list

# Separate soup object (one tome) into individual articles
# After the first tome, searching for 'div' becomes unreliable comapred to this one.
for passage in soup.find_all(type = 'article'): 
    
    # Try extracting bibliography text, don't fail if there isn't one.
    try:
         bibls.append(norm_and_strip(passage.bibl.getText()))
    except:
         bibls.append(None)
    # Extract titles
    titles.append(norm_and_strip(passage.head.getText()))

    # Extract article body text by combining everything after the bibliography-- catches ALL types of text tags
    # Try statement for the two cases in which there wasn't a bibliography-- in which combine the text after the title
    try:
        # Normalize, combine, and strip body text
        articles.append(" ".join([norm_and_strip(x.getText()) for x in passage.bibl.findNextSiblings()]))
    except:
        # Normalize, combine, and strip body text
        articles.append(" ".join([norm_and_strip(x.getText()) for x in passage.head.findNextSiblings()]))

# Remove puncutation from titles
for title in titles:
    capture = re.search(r'\[(.+)\]', title)
    if capture != None:
        i = titles.index(title)
        titles = titles[:i] + [capture.group(1)] + titles[i+1:]

# Create dictionary object which contains the extracted texts
dict = {'Bibliographies': bibls, 'Titles': titles, 'Articles': articles}

# Make DataFrame from dictionary
df = pd.DataFrame(dict)

# Write DataFrame to CSV for analysis in another notebook
df.to_csv('parsed_tomes_next_siblings.csv', index = False)

# Check CSV
pd.read_csv('parsed_tomes_next_siblings.csv').head()
```

```python
# All subsequent tomes will be written to CSV as 'appended'
# For loop to extract data from all tomes besides first
for tome in folder_files[1:]:
    
    # Read tome XML and parse into soup object
    with open('mercure-galant/xml/'+tome, 'r') as f:
        soup = bs.BeautifulSoup(f, 'xml')
    
    bibls = [] # Bibliography list
    titles = [] # Title list
    articles = [] # Article list
    
    # Separate soup object (one tome) into individual articles
    for passage in soup.find_all(type = 'article'):
        
        # Try extracting bibliography text without failing if none exists
        try:
            bibls.append(norm_and_strip(passage.bibl.getText()))
        except:
            bibls.append(None)
        # Extract title text
        titles.append(norm_and_strip(passage.head.getText()))
        # Try extracting article body text-- all tags which follow either bibliography or title if no bib exists
        try:
            articles.append(" ".join([norm_and_strip(x.getText()) for x in passage.bibl.findNextSiblings()]))
        except:
            articles.append(" ".join([norm_and_strip(x.getText()) for x in passage.head.findNextSiblings()]))
    
    # Remove punctuation from titles
    for title in titles:
            capture = re.search(r'\[(.+)\]', title)
            if capture != None:
                i = titles.index(title)
                titles = titles[:i] + [capture.group(1)] + titles[i+1:]
    # Create dictionary object which contains the extracted texts
    dict = {'Bibliographies': bibls, 'Titles': titles, 'Articles': articles}

    # Make DataFrame from dictionary
    df = pd.DataFrame(dict)
    
    # Write DataFrame to existing CSV in append mode and without headers to collect tomes 2 -> end along with tome 1
    df.to_csv('parsed_tomes_next_siblings.csv', mode = 'a', header = False, index = False)

# Check CSV
pd.read_csv('parsed_tomes_next_siblings.csv').head(10)
```
